```python
import pandas as pd
```

```python
kq_data = pd.read_csv("kenya_airways.csv")
kq_data.head()
```

```python
kq_data.info()
```

```python
kq_data['trip_verified'].unique()
```

```python
# get origin and destination
def split_flight_info(flight_info):
    parts = flight_info.split(" to ")
    origin = parts[0]
    destination = parts[1].split(" via ")[0] if " via " in parts[1] else parts[1]
    return pd.Series([origin, destination])

# Apply the function to split the route column
kq_data[['origin', 'destination']] = kq_data['route'].apply(split_flight_info)
kq_data.drop(columns=['route'], inplace=True)
kq_data.head()
```
